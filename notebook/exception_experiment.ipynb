```python
import sys
import traceback
print("hi")
```

```python
# Exception Handling
try:
    a = 9/0
    print(a)
except Exception as e:
    _,_,trc_back=sys.exc_info()
    file_name=trc_back.tb_frame.f_code.co_filename
    print(trc_back)
    print(file_name)
    lineno=trc_back.tb_lineno
    print(lineno)
    traceback_str = ''.join(traceback.format_exception(*sys.exc_info())) 
    print(traceback_str)
```
